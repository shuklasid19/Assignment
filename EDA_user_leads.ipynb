```python
#importing date and time libarary
import datetime 
from datetime import date
df["created_on"] = pd.datetime(df["created_on"])
```

```python
pd.to_datetime(df['created_on']).dt.time
```

```python
#extracting date and time
df['create_on_date'] = pd.to_datetime(df['created_on']).dt.date
df['created_on_Time'] = pd.to_datetime(df['created_on']).dt.time

df['updated_on_date'] = pd.to_datetime(df['updated_on']).dt.date
df['updated_on_Time'] = pd.to_datetime(df['updated_on']).dt.time
```

```python
#dropping original features 
df.drop(['created_on', 'updated_on'], axis=1, inplace=True)
```

```python
#creating features by extracting year, month , day from create_on feature 
df['create_on_date_year'] = pd.DatetimeIndex(df['create_on_date']).year
df['create_on_date_month'] = pd.DatetimeIndex(df['create_on_date']).month
df['create_on_date_day'] = pd.DatetimeIndex(df['create_on_date']).day
```

```python
##creating features by extracting year, month , day from updated_on feature 
df['updated_on_date_year'] = pd.DatetimeIndex(df['updated_on_date']).year
df['updated_on_date_month'] = pd.DatetimeIndex(df['updated_on_date']).month
df['updated_on_date_day'] = pd.DatetimeIndex(df['updated_on_date']).day
```

```python
#the difference between dates
df['difference_btw'] = df['create_on_date']-df['updated_on_date']
```
